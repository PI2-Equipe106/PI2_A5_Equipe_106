```python
import requests
import pandas as pd
```

```python
df_stations = pd.read_csv("C:/Users/alban/Downloads/donnees_stations_hydrometrie.csv",sep=",")
```

```python
for j in range(len(df_stations)):
    code_station = df_stations["code_station"][j]

    # Extraction du débit
    elab_url = "https://hubeau.eaufrance.fr/api/v1/hydrometrie/obs_elab"
    
    headers = {
        "Authorization": "Bearer VOTRE_TOKEN"  # Assurez-vous d'ajouter votre jeton d'authentification si nécessaire
    }

    all_results = []
    cursor = None  # Initialisez le curseur à None pour la première requête
    i = 1

    while True:
        params = {
            "code_entite": code_station,
            "cursor": cursor
        }

        response = requests.get(elab_url, headers=headers, params=params)

        if response.status_code == 200:
            data = response.json()
            results = data.get("data", [])
            all_results.extend(results)

            next_cursor = data.get("next", None)

            if next_cursor:
                print("Page number", i)
                i += 1
                cursor = next_cursor
            else:
                break  # Plus de pages suivantes, nous avons obtenu toutes les données

        else:
            print("Échec de la requête. Statut :", response.status_code)
            break

    # Créez un DataFrame à partir des données
    df_elab = pd.DataFrame(all_results)

    # Enregistrez le DataFrame au format CSV s'il contient des lignes
    if not df_elab.empty:
        path = "C:/Users/alban/OneDrive/Bureau/data_stations_elab/donnees_elab_sites_hydrometrie_station_" + code_station + ".csv"
        df_elab.to_csv(path, index=False)  # Le paramètre index=False évite d'inclure les numéros de ligne
```
