```python
merge_result['diff_prob_90'] = merge_result['prob_90'] - merge_result['probability_90_%']
merge_result['diff_prob_85'] = merge_result['prob_85'] - merge_result['probability_85_%']
merge_result['diff_prob_95'] = merge_result['prob_95'] - merge_result['probability_95_%']
```

```python
merge_result
```

```python
avg_abs_diff_90 = merge_result['diff_prob_90'].abs().mean()
avg_abs_diff_85 = merge_result['diff_prob_85'].abs().mean()
avg_abs_diff_80 = merge_result['diff_prob_95'].abs().mean()

print(f"Moyenne de la valeur absolue de diff_prob_90: {avg_abs_diff_90}")
print(f"Moyenne de la valeur absolue de diff_prob_85: {avg_abs_diff_85}")
print(f"Moyenne de la valeur absolue de diff_prob_95: {avg_abs_diff_80}")
```

```python
sorted_result = merge_result.sort_values(by='diff_prob_90', ascending=False)
```

```python
sorted_result
```

```python
merge_result.dropna(inplace=True)
```

```python
import matplotlib.pyplot as plt
import numpy as np


probabilities = ['90', '85']

for prob in probabilities:
    plt.figure(figsize=(15, 7))

    # SÃ©lection des donnÃ©es
    stations = merge_result['station']
    prob_col = 'prob_' + prob
    prob_perc_col = 'probability_' + prob + '_%'

    # Calcul de l'espacement pour chaque station
    bar_width = 0.35
    index = np.arange(len(stations))

    # Tracer les barres
    plt.bar(index, merge_result[prob_col], bar_width, label=f'prob_{prob}', color='b', alpha=0.6)
    plt.bar(index + bar_width, merge_result[prob_perc_col], bar_width, label=f'probability_{prob}_%', color='r', alpha=0.6)

    # Ajouter des titres et des lÃ©gendes
    plt.title(f'Comparaison des probabilitÃ©s {prob} pour chaque station')
    plt.xlabel('Stations')
    plt.ylabel(f'ProbabilitÃ© {prob}')
    plt.xticks(index + bar_width / 2, stations, rotation=90)
    plt.legend()

    # Afficher le plot
    plt.tight_layout()
    plt.show()
```
