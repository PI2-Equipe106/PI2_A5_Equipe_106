```python
import pandas as pd
import geopandas as gpd
import os

# Chemin du répertoire contenant les fichiers .shp
repertoire = "C:/Users/alban/OneDrive/Bureau/HYDRO/R84/HYDROGRAPHIE"

# Liste pour stocker les GeoDataFrames de chaque fichier .shp
gdfs = []

# Parcours de tous les fichiers .shp dans le répertoire
for fichier in os.listdir(repertoire):
    if fichier.endswith(".shp"):
        # Construction du chemin complet du fichier .shp
        chemin_fichier = os.path.join(repertoire, fichier)

        # Lecture du fichier .shp et ajout du GeoDataFrame à la liste
        gdf = gpd.read_file(chemin_fichier)
        gdfs.append(gdf)

# Concaténation de tous les GeoDataFrames en un seul
df_final = pd.concat(gdfs, ignore_index=True)
```

```python
# Enregistrement au format CSV
csv_path = 'C:/Users/alban/Downloads/Hydro_R84.csv'
df_final.to_csv(csv_path, index=False)
```
