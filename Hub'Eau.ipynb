```python
import requests
import pandas as pd

url = "https://hubeau.eaufrance.fr/api/v1/hydrometrie/obs_elab?cursor=&size=1000&date_debut_obs_elab=2005-01-01"

headers = {
    "Authorization": "Bearer VOTRE_TOKEN"  # Assurez-vous d'ajouter votre jeton d'authentification si ncessaire
}

all_results = []
i=1
while True:
    response = requests.get(url, headers=headers)

    if response.status_code == 206:
        data = response.json()
        results = data.get("data", [])
        all_results.extend(results)

        next_cursor = data.get("next", None)

        if next_cursor:
            print("page number",i)
            i=i+1
            url = next_cursor
        else:
            break  # Plus de pages suivantes, nous avons obtenu toutes les donnes

    else:
        print("chec de la requte. Statut :", response.status_code)
        break
```

```python
# Crez un DataFrame  partir des donnes
df = pd.DataFrame(all_results)
```

```python
print(len(df))
```

```python
# Enregistrez le DataFrame au format CSV
df.to_csv("C:/Users/alban/Downloads/donnees_hydrometrie.csv", index=False)  # Le paramtre index=False vite d'inclure les numros de ligne
```
