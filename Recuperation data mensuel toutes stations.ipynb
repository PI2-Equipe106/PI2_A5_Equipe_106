```python
import requests
import pandas as pd
```

```python
df_stations = pd.read_csv("C:/Users/alban/Downloads/donnees_stations_hydrometrie.csv",sep=",")
```

```python
for j in range(len(df_stations)):
    code_station = df_stations["code_station"][j]

    # Extraction du débit
    debit_url = "https://hubeau.eaufrance.fr/api/v1/hydrometrie/observations_tr?code_entite=" + code_station + "&size=10000&pretty&grandeur_hydro=Q&fields=code_station,date_debut_serie,date_fin_serie,date_obs,resultat_obs,continuite_obs_hydro"

    headers = {
        "Authorization": "Bearer VOTRE_TOKEN"  # Assurez-vous d'ajouter votre jeton d'authentification si nécessaire
    }

    all_results = []
    i = 1
    while True:
        response = requests.get(debit_url, headers=headers)

        if response.status_code == 200:
            data = response.json()
            results = data.get("data", [])
            all_results.extend(results)

            next_cursor = data.get("next", None)

            if next_cursor:
                print("page number", i)
                i = i + 1
                debit_url = next_cursor
            else:
                break  # Plus de pages suivantes, nous avons obtenu toutes les données

        else:
            print("Échec de la requête. Statut :", response.status_code)
            break

    # Créez un DataFrame à partir des données
    df_debit = pd.DataFrame(all_results)

    # Enregistrez le DataFrame au format CSV s'il contient des lignes
    if not df_debit.empty:
        path = "C:/Users/alban/OneDrive/Bureau/data_stations_debit/donnees_sites_hydrometrie_station_" + code_station + ".csv"
        df_debit.to_csv(path, index=False)  # Le paramètre index=False évite d'inclure les numéros de ligne
```

```python
for j in range(len(df_stations)):
    code_station = df_stations["code_station"][j]
    # Extraction de la hauteur
    hauteur_url = "https://hubeau.eaufrance.fr/api/v1/hydrometrie/observations_tr?code_entite=" + code_station + "&size=10000&pretty&grandeur_hydro=H&fields=code_station,date_debut_serie,date_fin_serie,date_obs,resultat_obs,continuite_obs_hydro"

    all_results = []
    i = 1
    while True:
        response = requests.get(hauteur_url, headers=headers)

        if response.status_code == 200:
            data = response.json()
            results = data.get("data", [])
            all_results.extend(results)

            next_cursor = data.get("next", None)

            if next_cursor:
                print("page number", i)
                i = i + 1
                hauteur_url = next_cursor
            else:
                break  # Plus de pages suivantes, nous avons obtenu toutes les données

        else:
            print("Échec de la requête. Statut :", response.status_code)
            break

    # Créez un DataFrame à partir des données
    df_hauteur = pd.DataFrame(all_results)

    # Enregistrez le DataFrame au format CSV s'il contient des lignes
    if not df_hauteur.empty:
        path = "C:/Users/alban/OneDrive/Bureau/data_stations_hauteur/donnees_sites_hydrometrie_station_" + code_station + ".csv"
        df_hauteur.to_csv(path, index=False)  # Le paramètre index=False évite d'inclure les numéros de ligne

print("Done")
```
