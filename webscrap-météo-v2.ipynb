```python
#Notebook final avec noms traités

# TRAITEMENT DES NOMS DE STATIONS : 
# enlever majuscule, remplacer espaces par des -, supprimer accents, gestion apostrophes, points, ...


from bs4 import BeautifulSoup
import requests
from unidecode import unidecode
import re

# URL de la page contenant les stations
url = "https://www.infoclimat.fr/stations-meteo/analyses-mensuelles.php?mois=9&annee=2023"

# Envoyer une requête GET pour récupérer le contenu de la page
response = requests.get(url)

# Vérifier si la requête a réussi (statut 200 OK)
if response.status_code == 200:
    # Utiliser BeautifulSoup pour analyser le contenu HTML de la page
    soup = BeautifulSoup(response.content, 'html.parser')
    
    # Trouver le sélecteur de station
    station_select = soup.find('select', {'name': 'staid'})
    
    if station_select:
        # Extraire les options (stations) de la liste déroulante
        station_options = station_select.find_all('option')
        
        # Créer une liste pour stocker les stations après traitement
        stations = []
        
        # Parcourir les options pour extraire les stations
        for option in station_options:
            station_value = option['value']
            station_name = option.get_text(strip=True)
            
            # Enlever le département entre parenthèses
            station_name = station_name.split('(')[0].strip()
            
            # Remplace les tirets par un espace
            station_name = station_name.replace("-", " ")
            
            # Remplace les apostrophes par un espace
            station_name = station_name.replace("'", " ")
            
            # Remplace les slash par un espace
            station_name = station_name.replace("/", " ")
            
            # Remplace les & par rien
            station_name = station_name.replace("&", "")
            
            # Remplace les espaces multiples par un seul espace
            station_name = re.sub(r'\s+', ' ', station_name)
            
            # Supprimer les espaces et convertir en minuscules
            station_name = station_name.replace(" ", "-").lower()
            
            # Supprimer les accents
            station_name = unidecode(station_name)
            
            # Remplace les points par une chaîne vide
            station_name = station_name.replace(".", "")
            
            # Vérifie si le dernier caractère est un tiret et le supprime si c'est le cas
            if station_name[-1] == '-':
                station_name = station_name[:-1]
            
            stations.append((station_value, station_name))
        
        # Afficher la liste des stations (format : (value, name))
        for station in stations:
            print(f"Value: {station[0]}, Name: {station[1]}")
    else:
        print("Sélecteur de station non trouvé sur la page.")
else:
    print("Échec de la requête. Statut :", response.status_code)
```
